```python
import numpy as np
from sklearn.datasets import load_breast_cancer
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

d = load_breast_cancer()

data = pd.DataFrame(d.data, columns=d.feature_names)
target = pd.DataFrame(d.target)

train_X, test_X, train_y, test_y = train_test_split(data, target, test_size=0.3, random_state=2019, shuffle=True)
```

```python
d = (train_X - np.mean(train_X))/np.std(train_X)
data = d
test_X = (test_X - np.mean(test_X)) / np.std(test_X)
t = test_X
```

```python
d['const'] = np.ones_like(d['mean area'])
t['const'] = np.ones_like(test_X['mean area'])
```

```python
def logit(x, y, p=False, lr=0.001):
#     if not p:
#         p = np.random.normal(0, 1, len(x[0]))
    t = np.array(list(map(lambda x: 1/(1 + np.exp(x)), x@p))) - y.flatten()
    
    res = t.T @ x
    return p + np.dot(res, lr/len(x)), np.sum(res)/len(x)
```

```python
np.random.seed(0)
p = np.random.normal(0, 0.1, len(d.values[0]))
# p = np.zeros_like(d.columns)
for _ in range(200):
    p, j = logit(d.values, train_y.values, p, lr=0.01)
    print(_, '######')
    print(j)
    print(p)
    if j < 0:
        break
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0, penalty='none').fit(d, train_y)
```

```python
accuracy_score(test_y, np.array(list(map(lambda x: 1 if 1/(1+x) > 0 else 0, np.dot(t.values @ p, -1)))))
```

```python
accuracy_score(test_y, lr.predict(test_X))
```
