```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler

d = load_boston()

std = StandardScaler()
data = pd.DataFrame(std.fit_transform(d.data), columns=d.feature_names)
target = std.fit_transform(d.target.reshape(-1,1))
```

```python
data
```

```python
data['const'] = np.ones_like(data.iloc[:,0])
```

```python
class Percep:
    def __init__(self, x, y, w = None, activation='linear'):
        np.random.seed(0)
        self.data = x
        self.target = y
        self.w = np.random.normal(0, 1, len(x.values[0]))
        self.activation = activation
        self.loss = np.Inf
        self.hist = []
        
    def update(self, x, y, w, lr=0.01):
        x = x.values
        y = y.flatten()
        res = x.T @ (y - np.dot(x, w))
        
        if self.activation == 'relu':
            return w - np.dot(res, -lr/len(x)), (y - np.dot(x, w)).T @ (y - np.dot(x, w)) / len(x)
        return w - np.dot(res, -lr/len(x)), (y - np.dot(x, w)).T @ (y - np.dot(x, w)) / len(x) 
        
    def fit(self):
        for _ in range(200):
            w, c = self.update(self.data, self.target, self.w)
            self.hist.append(c)
            if np.abs(c-self.loss) < 1e-4:
                break
            else:
                self.w = w
                self.loss = c
                       
```

```python
nn = Percep(data, target)
```

```python
nn.fit()
```
